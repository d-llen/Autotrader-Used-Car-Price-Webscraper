### functions

```python
def scrape_autotrader(cars, criteria, name):
    
    def check_temp_block(html_text):
        """Detect Auto Trader temp block or CAPTCHA page."""
        block_indicators = [
            "unusual activity",
            "verify you are a human",
            "captcha",
            "access to this page has been denied",
            "temporarily blocked",
            "are you a human",
        ]
        html_lower = (html_text or "").lower()
        return any(ind in html_lower for ind in block_indicators)

    def safe_get_page(driver, url, sleep_sec=random.uniform(4,6)):
        """Navigate and return page_source safely."""
        try:
            driver.get(url)
            time.sleep(sleep_sec)
            return driver.page_source
        except Exception as e:
            print(f"[warn] Navigation error for {url[:120]}... :: {e}")
            return ""

    chrome_options = Options()
    chrome_options.add_argument("_tt_enable_cookie=1")
    driver = None
    data = []

    # Track for debugging / resilience
    pages_scraped_this_car = 0

    try:
        driver = webdriver.Chrome(options=chrome_options)

        for car in cars:
            start = time.time()
            print("\n")

            # Build params safely (no empty keys)
            params = []
            try:
                if car.get("make"):
                    params.append(f"make={car['make']}")
                if car.get("model"):
                    params.append(f"model={car['model']}")
                if criteria.get("postcode"):
                    params.append(f"postcode={criteria['postcode']}")
                if criteria.get("radius"):
                    params.append(f"radius={criteria['radius']}")
                params.append("sort=relevance")
                if criteria.get("year_from"):
                    params.append(f"year-from={criteria['year_from']}")
                if criteria.get("year_to"):
                    params.append(f"year-to={criteria['year_to']}")
                if criteria.get("price_from"):
                    params.append(f"price-from={criteria['price_from']}")
                if criteria.get("price_to"):
                    params.append(f"price-to={criteria['price_to']}")
            except Exception as e:
                print(f"[warn] Param build failed for car {car} :: {e}")
                continue

            base = "https://www.autotrader.co.uk/car-search?advertising-location=at_cars&include-delivery-option=on&"
            url = base + "&".join(params)

            # Initial page
            source = safe_get_page(driver, url, sleep_sec=5)
            if not source:
                print("[warn] Empty initial page source, skipping car.")
                continue

            if check_temp_block(source):
                print("\n[!] TEMP BLOCK DETECTED — exporting data collected so far...")
                return export(data, criteria, name)

            content = BeautifulSoup(source, "html.parser")

            # Count pages safely
            try:
                results = content.find("span", attrs={"data-testid": "search-result-count"})
                if not results:
                    print("[warn] No results span found; defaulting to 1 page.")
                    number_of_pages = 1
                    number_of_articles = "unknown"
                else:
                    number_of_articles = results.text.split()[0].replace(",", "")
                    number_of_pages = int(number_of_articles) // 19
                    if number_of_pages == 0:
                        number_of_pages = 1
                    if number_of_pages > 100:
                        number_of_pages = 100
                        number_of_articles = "approx. 2600"
            except Exception as e:
                print(f"[warn] Could not determine pages :: {e}; defaulting to 1 page.")
                number_of_pages, number_of_articles = 1, "unknown"

            try:
                if car.get("make"):
                    print(f"Manufacturer: {car['make']}")
                    if car.get("model"):
                        print(f"Model: {car['model']}")
                print(f"There are {number_of_pages} pages in total for {number_of_articles} cars.")
                ETAseconds = int(number_of_pages) * 11
                ETAminutes = ETAseconds / 60
                if ETAseconds < 120:
                    print(f"Estimated search time: {ETAseconds} seconds.")
                else:
                    print(f"Estimated search time: {ETAminutes:.1f} minutes.")
            except Exception:
                pass  # non-critical printing

            # Paginated pages
            pages_scraped_this_car = 0
            for i in range(int(number_of_pages)):
                page_url = url + f"&page={str(i + 1)}"
                page_source = safe_get_page(driver, page_url, sleep_sec=5)

                if check_temp_block(page_source):
                    print("\n[!] TEMP BLOCK DETECTED mid-run — exporting data collected so far...")
                    return export(data, criteria, name)

                if not page_source:
                    print(f"[warn] Empty page at i={i+1}; continuing.")
                    continue

                content = BeautifulSoup(page_source, "html.parser")

                try:
                    articles = content.findAll("div", attrs={"data-testid": "advertCard"})
                except Exception as e:
                    print(f"[warn] Could not find articles on page {i+1} :: {e}")
                    articles = []

                # If the HTML changed (e.g., cookie wall) and no articles found several times, bail this car.
                if not articles and i > 0 and pages_scraped_this_car == 0:
                    print("[warn] No articles found on first paginated page; moving to next car.")
                    break

                for article in articles:
                    details = {
                        "name": None,
                        "price": None,
                        "year": None,
                        "mileage": None,
                        "transmission": None,
                        "fuel": None,
                        "engine": None,
                        "owners": None,
                        "location": None,
                        "distance": None,
                        "link": None,
                    }

                    # Everything inside a try so one bad card doesn't kill the page
                    try:
                        # Name
                        if car.get("make"):
                            details["make"] = car["make"]
                            if car.get("model"):
                                details["model"] += car["model"]
                            else: 
                                details["model"] = "Unspecified"

                        # Price
                        try:
                            price_match = re.search(r"[£]\d+(?:,\d{3})*", article.get_text(" ", strip=True))
                            if price_match:
                                details["price"] = price_match.group(0)
                        except Exception:
                            pass

                        # Link
                        try:
                            link_tag = article.find("a", href=re.compile(r"/car-details/"))
                            if link_tag and link_tag.has_attr("href"):
                                details["link"] = link_tag["href"]
                        except Exception:
                            pass

                        # Location & distance (robust)
                        try:
                            raw = article.find("span", attrs={"data-testid": "search-listing-location"})
                            if raw:
                                text = " ".join(raw.get_text(" ", strip=True).split())
                                text = re.sub(r"(?i)dealer location[:\s]*", "", text).strip()
                                m = re.search(r"(\d+)\s*miles?", text, flags=re.I)
                                details["distance"] = m.group(1) if m else None
                                loc = re.sub(r"\(\s*[^)]*\)", "", text)  # remove (... )
                                loc = re.sub(r"\d+\s*miles?", "", loc, flags=re.I)
                                details["location"] = loc.strip(" ,.-") or None
                        except Exception:
                            pass

                        # Specs
                        try:
                            specs_list = article.find("ul", attrs={"data-testid": "badges-container"})
                            if specs_list:
                                for spec in specs_list.find_all("li"):
                                    text = spec.get_text(strip=True)
                                    if "reg" in text:
                                        details["year"] = text
                                    if "miles" in text:
                                        details["mileage"] = text
                                    if text in ["Manual", "Automatic"]:
                                        details["transmission"] = text
                                    if "." in text and "L" in text:
                                        details["engine"] = text
                                    if text in ["Petrol", "Diesel", "Hybrid", "Electric"]:
                                        details["fuel"] = text
                                    if "owner" in text.lower():
                                        # e.g. "2 owners"
                                        om = re.search(r"(\d+)\s*owner", text, flags=re.I)
                                        details["owners"] = om.group(1) if om else text[:1]
                        except Exception:
                            pass

                        data.append(details)

                    except Exception as e:
                        print(f"[warn] Failed to parse an article card :: {e}")
                        # keep going on the rest of the page

                pages_scraped_this_car += 1
                print(f"Page {str(i + 1)} scraped. ({len(articles)} articles)")
                # Light pacing already done by safe_get_page's sleep

            print("\n")
            end = time.time()
            print(f"Car done in {end - start:.2f}s")

        # Finished all cars normally
        return export(data, criteria, name)

    except KeyboardInterrupt:
        print("\n[!] KeyboardInterrupt — exporting partial data...")
        return export(data, criteria, name)
    except Exception as e:
        print(f"\n[!] Unhandled error: {e} — exporting partial data...")
        return export(data, criteria, name)
    finally:
        try:
            if driver:
                driver.quit()
        except Exception:
            pass


def export(data, criteria, name):
    import os, re, datetime
    import pandas as pd

    df = pd.DataFrame(data)

    # If totally empty, still create an empty CSV with headers to avoid breaking pipelines
    if df.empty:
        cols = ["make","model","link","price","year","mileage","miles_pa","owners","distance","location","engine","transmission","fuel"]
        df = pd.DataFrame(columns=cols)

    # Cleaning — resilient to weird inputs
    if "price" in df:
        df["price"] = (df["price"].astype(str)
                       .str.replace("£", "", regex=False)
                       .str.replace(",", "", regex=False))
        df["price"] = pd.to_numeric(df["price"], errors="coerce").astype("Int64")

    if "year" in df:
        df["year"] = df["year"].astype(str).str.replace(r"\s(\(\d\d reg\))", "", regex=True)
        df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")

    if "mileage" in df:
        df["mileage"] = (df["mileage"].astype(str)
                         .str.replace(",", "", regex=False)
                         .str.replace(" miles", "", regex=False))
        df["mileage"] = pd.to_numeric(df["mileage"], errors="coerce").astype("Int64")

    now = datetime.datetime.now()
    if "mileage" in df and "year" in df:
        # avoid division by zero / NaN
        age = (now.year - df["year"]).replace(0, pd.NA)
        df["miles_pa"] = pd.to_numeric(df["mileage"], errors="coerce") / age
        df["miles_pa"] = df["miles_pa"].fillna(0).astype(int)
    else:
        df["miles_pa"] = 0

    if "owners" in df:
        df["owners"] = pd.to_numeric(df["owners"], errors="coerce").fillna(-1).astype(int)
    else:
        df["owners"] = -1

    if "distance" in df:
        df["distance"] = pd.to_numeric(df["distance"], errors="coerce").fillna(-1).astype(int)
    else:
        df["distance"] = -1

    if "link" in df:
        df["link"] = "https://www.autotrader.co.uk" + df["link"].astype(str)

    # Ensure all expected columns exist
    expected = ["make","model","link","price","year","mileage","miles_pa","owners","distance","location","engine","transmission","fuel"]
    for col in expected:
        if col not in df.columns:
            df[col] = pd.NA

    df = df[expected]

    if criteria.get("price_to"):
        try:
            df = df[df["price"] < int(criteria["price_to"])]
        except Exception:
            pass

    try:
        df = df.sort_values(by="distance", ascending=True)
    except Exception:
        pass

    # Output path
    now_str = now.strftime("%d-%m-%Y")
    folder_path = os.path.join("scraped_data", now_str)
    os.makedirs(folder_path, exist_ok=True)

    # Sanitize filename and add time HHMM to avoid overwrites
    def sanitize(s: str) -> str:
        return re.sub(r'[<>:"/\\|?*\n\r\t]+', "_", str(s)).strip()

    name = f"{sanitize(name)}_{now.strftime('%H%M')}.csv"
    file_path = os.path.join(folder_path, name)
    try:
        df.to_csv(file_path, index=False)
        print(f"[+] Data saved to {file_path} ({len(df)} rows)")
    except Exception as e:
        print(f"[warn] Could not save CSV ({e}). Returning DataFrame anyway.")

    return df
```
